# Quesiton 1 Explain the theoretical Simple Linear Regression model in your own words

In a theoretical Simple Linear Regression (SLR) model, we aim to explain or predict the outcome variable, ùëå , using a predictor variable, ùëã . The model assumes a linear relationship between ùëã  and ùëå  

Y=Œ≤0 ‚Äã +Œ≤1 ‚Äã X+Œµ where: 

Y is the outcome (or dependent) variable. 

X is the predictor (or independent) variable.

Œ≤ 0 ‚Äã is the intercept, representing the value of Y when ùëã = 0 .

ùõΩ 1 ‚Äã is the slope, indicating the change in Y for a one-unit increase in X. 

 Œµ is the error term, which accounts for the variation in Y not explained by X.
 
 #### consider the following code to better help understand

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the SLR model
beta_0 = 5       # Intercept
beta_1 = 2       # Slope
sigma = 3        # Standard deviation of the error term

# Generating a sample of predictor variable X
np.random.seed(0)
n = 100          # Number of observations
X = np.random.uniform(0, 10, n)

# Generating the outcome variable Y based on the model Y = beta_0 + beta_1 * X + epsilon
epsilon = np.random.normal(0, sigma, n)
Y = beta_0 + beta_1 * X + epsilon

# Plotting the data and the theoretical regression line
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', alpha=0.6, label="Sample Data")
plt.plot(X, beta_0 + beta_1 * X, color='red', label="True Regression Line")

# Show normal distributions of Y around the regression line for a few X values
X_values = [2, 5, 8]  # Chosen X values to illustrate distributions
for x_val in X_values:
    y_mean = beta_0 + beta_1 * x_val
    y_distribution = norm(loc=y_mean, scale=sigma)
    y_range = np.linspace(y_mean - 3*sigma, y_mean + 3*sigma, 100)
    plt.plot([x_val]*100, y_range, 'k--', alpha=0.6)  # Vertical line at X
    plt.plot(x_val, y_mean, 'ro')  # Mean of Y for chosen X value
    plt.fill_betweenx(y_range, x_val - 0.2, x_val + 0.2, alpha=0.2, color='gray')
    plt.text(x_val, y_mean + 1, f"$N({y_mean:.1f}, {sigma**2})$", ha='center')

# Labels and legend
plt.xlabel("Predictor Variable X")
plt.ylabel("Outcome Variable Y")
plt.legend()
plt.title("Simple Linear Regression Model with Normal Distributions of Y at Selected X Values")
plt.show()
```

# Question 2 visualize a fitted Simple Linear Regression model using pandas and import statsmodels.formula.api as smf

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Parameters for the theoretical Simple Linear Regression model
beta_0 = 5        # Intercept
beta_1 = 2        # Slope
sigma = 3         # Standard deviation of the error term

# Simulating the dataset
np.random.seed(0)
n = 100          # Number of observations
X = np.random.uniform(0, 10, n)  # Predictor variable
epsilon = np.random.normal(0, sigma, n)  # Error term
Y = beta_0 + beta_1 * X + epsilon  # Outcome variable

# Creating a DataFrame
data = pd.DataFrame({'X': X, 'Y': Y})

# Fit a Simple Linear Regression model using statsmodels
model = smf.ols(formula='Y ~ X', data=data)
results = model.fit()

# Print the summary of the model
print(results.summary())

# Visualizing the data and the fitted line
plt.figure(figsize=(10, 6))
plt.scatter(data['X'], data['Y'], color='blue', alpha=0.6, label="Simulated Data")
plt.plot(data['X'], beta_0 + beta_1 * data['X'], color='red', linestyle='--', label="True Line")
plt.plot(data['X'], results.fittedvalues, color='green', label="Fitted Regression Line")

# Labels and legend
plt.xlabel("Predictor Variable X")
plt.ylabel("Outcome Variable Y")
plt.legend()
plt.title("Fitted Simple Linear Regression Model")
plt.show()
```

# Question 4 Explain how fitted_model.fittedvalues are derived on the basis of fitted_model.summary().tables[1]

Illustration of How fitted_model.fittedvalues are Computed
Let's manually calculate the predicted values using fitted_model.params for a clearer illustration:


```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf

# Parameters for the theoretical Simple Linear Regression model
beta_0 = 5        # Intercept (True value)
beta_1 = 2        # Slope (True value)
sigma = 3         # Standard deviation of the error term

# Simulating the dataset
np.random.seed(0)
n = 100          # Number of observations
X = np.random.uniform(0, 10, n)  # Predictor variable
epsilon = np.random.normal(0, sigma, n)  # Error term
Y = beta_0 + beta_1 * X + epsilon  # Outcome variable

# Creating a DataFrame
data = pd.DataFrame({'X': X, 'Y': Y})

# Fit a Simple Linear Regression model using statsmodels
fitted_model = smf.ols(formula='Y ~ X', data=data).fit()

# Print the summary of the model to see fitted coefficients
print(fitted_model.summary())

# Extracting the estimated coefficients
intercept = fitted_model.params['Intercept']
slope = fitted_model.params['X']

# Manually calculating the fitted values using the estimated parameters
data['predicted_Y'] = intercept + slope * data['X']

# Comparing the manually calculated predictions with fitted_model.fittedvalues
print("First 5 Manual Predictions:", data['predicted_Y'].head())
print("First 5 fitted_model.fittedvalues:", fitted_model.fittedvalues.head())

```

### Derivation of fitted_model.fittedvalues The fitted_model.fittedvalues represents the "in-sample predictions" or predicted ùëå-values for each observed ùëã in the dataset, based on the estimated model.

In summary: The theoretical model uses known parameters ( ùõΩ 0  and ùõΩ 1  ‚Äã ) to predict ùëå . The fitted model uses estimated parameters ( ùõΩ^ 0  and ùõΩ^ 1 ) to predict ùëå based on the actual sample data, yielding fitted_model.fittedvalues.

 Null Hypothesis Specification For a Simple Linear Regression model of the form: ùëå = ùõΩ 0 + ùõΩ 1 ùëã + ùúñ  where ùëå  represents the eruption duration and ùëã X represents the waiting time, we specify the null and alternative hypotheses as follows: Null Hypothesis ( ùêª 0 ): There is no linear association between waiting time and eruption duration, which implies that the slope parameter ùõΩ 1 = 0. Alternative Hypothesis ( ùêª 1  ): There is a linear association between waiting time and eruption duration, implying ùõΩ 1 ‚â† 0

If the p-value for ùõΩ 1  is low, we would reject the null hypothesis and conclude there is evidence of a significant linear association between waiting time and eruption duration. Conversely, if the p-value is high, we would fail to reject the null hypothesis, suggesting insufficient evidence for a linear relationship. This p-value-based interpretation allows us to determine if the observed association is likely due to chance or reflects a genuine relationship in the population.

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Load the Old Faithful Geyser dataset
old_faithful = sns.load_dataset('geyser')

# Step 1: Filter for long wait times and take a sample of n=160
long_wait = old_faithful[old_faithful['waiting'] >= 70]
long_wait_sample = long_wait.sample(n=160, random_state=42)

# Bootstrap analysis
n_bootstrap = 1000
bootstrap_slopes = []

for _ in range(n_bootstrap):
    # Resample with replacement
    bootstrap_sample = long_wait_sample.sample(n=len(long_wait_sample), replace=True)
    
    # Fit the model to the bootstrap sample
    model = smf.ols('duration ~ waiting', data=bootstrap_sample).fit()
    
    # Collect the slope coefficient
    bootstrap_slopes.append(model.params['waiting'])

# Visualize the bootstrapped sampling distribution of slope coefficients
plt.figure(figsize=(10, 6))
plt.hist(bootstrap_slopes, bins=30, edgecolor='k', alpha=0.7)
plt.axvline(0, color='red', linestyle='--', label='Slope = 0')
plt.title('Bootstrapped Sampling Distribution of Slope Coefficient')
plt.xlabel('Slope Coefficient')
plt.ylabel('Frequency')
plt.legend()
plt.show()

```

We perform 1000 bootstrap samples with replacement from the long wait data subset.
For each bootstrap sample, we fit a linear regression model and collect the slope coefficient.
The histogram shows the distribution of these bootstrapped slope coefficients.

# 10b)

```python
from scipy.stats import norm

# Parameters for the null hypothesis simulation
B0 = 1.65  # Intercept
B1 = 0     # Null hypothesis slope
sigma = 0.37  # Standard deviation of residuals
n_simulations = 1000
simulated_slopes = []

# Simulate sampling distribution under null hypothesis
for _ in range(n_simulations):
    # Generate simulated duration values assuming B1 = 0
    waiting_times = long_wait_sample['waiting'].values
    simulated_duration = B0 + B1 * waiting_times + norm.rvs(scale=sigma, size=len(waiting_times))
    
    # Create a DataFrame for the simulated sample
    simulated_sample = pd.DataFrame({'waiting': waiting_times, 'duration': simulated_duration})
    
    # Fit the model to the simulated sample
    sim_model = smf.ols('duration ~ waiting', data=simulated_sample).fit()
    
    # Collect the slope coefficient
    simulated_slopes.append(sim_model.params['waiting'])

# Visualize the simulated sampling distribution of the slope coefficient
plt.figure(figsize=(10, 6))
plt.hist(simulated_slopes, bins=30, edgecolor='k', alpha=0.7)
plt.axvline(0, color='red', linestyle='--', label='Slope = 0')
plt.title('Simulated Sampling Distribution of Slope Coefficient (Null Hypothesis)')
plt.xlabel('Slope Coefficient')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

We simulate samples from a Simple Linear Regression model under the null hypothesis where ùõΩ 0 = 1.65 , ùõΩ 1 = 0 , and ùúé = 0.37 

# 10c)

```python
# Calculate the 95% confidence interval from bootstrapped slopes
ci_lower, ci_upper = np.percentile(bootstrap_slopes, [2.5, 97.5])
print(f"95% Bootstrapped Confidence Interval for slope: ({ci_lower:.4f}, {ci_upper:.4f})")

# Fit the model on the actual long wait sample to get the observed slope and p-value
fitted_model = smf.ols('duration ~ waiting', data=long_wait_sample).fit()
observed_slope = fitted_model.params['waiting']
actual_p_value = fitted_model.pvalues['waiting']
print(f"Observed slope from the actual data: {observed_slope:.4f}")
print(f"Actual p-value from the fitted model: {actual_p_value:.4f}")

# Calculate the simulated p-value under the null hypothesis
simulated_p_value = np.mean([abs(slope) >= abs(observed_slope) for slope in simulated_slopes])
print(f"Simulated p-value under null hypothesis: {simulated_p_value:.4f}")

# Report if 0 is within the confidence interval
if ci_lower <= 0 <= ci_upper:
    print("0 is contained within the 95% bootstrapped confidence interval, so we fail to reject the null hypothesis.")
else:
    print("0 is NOT contained within the 95% bootstrapped confidence interval, so we have evidence against the null hypothesis.")
```

```python
import numpy as np

# Define observed mean difference
observed_diff = old_faithful[short_wait]['duration'].mean() - old_faithful[long_wait]['duration'].mean()

# Number of permutations
n_permutations = 10000
permuted_diffs = []

# Combine data and perform permutation
combined_data = old_faithful['duration'].values
labels = np.concatenate([np.ones(len(old_faithful[short_wait])), np.zeros(len(old_faithful[long_wait]))])

for _ in range(n_permutations):
    # Shuffle the labels
    np.random.shuffle(labels)
    short_sample = combined_data[labels == 1]
    long_sample = combined_data[labels == 0]
    
    # Compute mean difference
    permuted_diff = np.mean(short_sample) - np.mean(long_sample)
    permuted_diffs.append(permuted_diff)

# Calculate p-value
p_value = np.mean(np.abs(permuted_diffs) >= np.abs(observed_diff))
print(f"Observed Mean Difference: {observed_diff:.4f}")
print(f"Permutation Test p-value: {p_value:.4f}")
```

```python
n_bootstrap = 10000
bootstrapped_mean_differences = []

for _ in range(n_bootstrap):
    short_sample = np.random.choice(old_faithful[short_wait]['duration'], size=len(old_faithful[short_wait]), replace=True)
    long_sample = np.random.choice(old_faithful[long_wait]['duration'], size=len(old_faithful[long_wait]), replace=True)
    
    # Calculate mean difference
    boot_mean_diff = np.mean(short_sample) - np.mean(long_sample)
    bootstrapped_mean_differences.append(boot_mean_diff)

# Calculate 95% CI
confidence_interval = np.quantile(bootstrapped_mean_differences, [0.025, 0.975])
print(f"95% Bootstrap Confidence Interval for Difference in Means: {confidence_interval}")
```
